```python
import tensorflow as tf
import random
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
x_train = x_train.reshape((60000, 28, 28, 1))

for i in range(4):
  random_index = random.randint(0, len(x_train))
  plt.imshow(x_train[random_index].reshape((28, 28)), cmap='gray')
  plt.show()
```

```python
#Image Negatives
def digital_negative(img):
  neg=[]
  for i in range(len(img)):
    temp=[]
    for j in range(len(img)):
      s=256-1-img[i][j]
      temp.append(s)
    neg.append(temp)
  return np.array(neg)
```

```python
def thresholding(img, threshold=127):
  new_img=img.copy()
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      if img[i][j]>threshold:
        new_img[i][j]=255
      else:
        new_img[i][j]=0
  return (new_img)
```

```python
def gls_without_bg(img,a,b):
  neg=[]
  for i in range(len(img)):
    temp=[]
    for j in range(len(img)):
      if(img[i][j]>=a and img[i][j]<=b):
        s=255
      else:
        s=0
      temp.append(s)
    neg.append(temp)
  return np.array(neg)

```

```python
def gls_with_bg(img,a,b):
  neg=np.zeros_like(img)
  for i in range(len(img)):
    for j in range(len(img)):
      if(img[i][j]>=a and img[i][j]<=b):
        neg[i][j] = img[i][j]
      else:
        neg[i][j] = 0
  return neg
```
