```python
img = x_train[0]
img1 = x_train[1]

add_img = cv.divide(img,img1)



fig, axs = plt.subplots(1, 3, figsize=(10, 5))


axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')  # Hide axes

axs[1].imshow(img1)
axs[1].set_title('Original Image')
axs[1].axis('off')

axs[2].imshow(add_img)
axs[2].set_title('Divided Image')
axs[2].axis('off')

plt.show()
```

```python
# Sample images from the dataset
img = x_train[0]
img1 = x_train[1]

# Convert images to float32 to handle blending properly
img = img.astype(np.float32)
img1 = img1.astype(np.float32)

# Define weights for blending
alpha = 0.7  # Weight for the first image
beta = 0.3   # Weight for the second image
gamma = 0    # Scalar added to each sum

# Perform weighted addition (blending)
blended_image = cv.addWeighted(img, alpha, img1, beta, gamma)

# Convert back to uint8
blended_image = np.clip(blended_image, 0, 255).astype(np.uint8)

# Set up the plotting
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Display the original images and the result of blending
axs[0].imshow(cv.cvtColor(img.astype(np.uint8), cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[0].set_title('Original Image 1')
axs[0].axis('off')

axs[1].imshow(cv.cvtColor(img1.astype(np.uint8), cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[1].set_title('Original Image 2')
axs[1].axis('off')

axs[2].imshow(cv.cvtColor(blended_image, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[2].set_title('Blended Image')
axs[2].axis('off')

# Show the plot
plt.show()
```

```python
img = x_train[0]
img1 = x_train[1]

# Convert images to float32 to handle subtraction properly
img = img.astype(np.float32)
img1 = img1.astype(np.float32)

# Define weights for subtraction
alpha = 0.7  # Weight for the first image
beta = 0.3   # Weight for the second image

subtracted_image = cv.subtract(img, img1 * beta)

subtracted_image = np.clip(subtracted_image, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv.cvtColor(img.astype(np.uint8), cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[0].set_title('Original Image 1')
axs[0].axis('off')

axs[1].imshow(cv.cvtColor(img1.astype(np.uint8), cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[1].set_title('Original Image 2')
axs[1].axis('off')

axs[2].imshow(cv.cvtColor(subtracted_image, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
axs[2].set_title('Subtracted Image')
axs[2].axis('off')
plt.show()
```
