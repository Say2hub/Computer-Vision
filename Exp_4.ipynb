```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
image = train_images[np.random.randint(0, train_images.shape[0])]
```

```python
# Step 1: Calculate the histogram of the image
histogram, bin_edges = np.histogram(image.flatten(), bins=256, range=[0, 256])
```

```python
# Step 2: Calculate the Probability Mass Function (PMF)
pmf = histogram / np.sum(histogram)
```

```python
# Step 3: Calculate the Cumulative Distribution Function (CDF)
cdf = pmf.cumsum()
```

```python
# Step 4: Find the new gray level values by multiplying the CDF with 255
new_gray_values = (cdf * 255).astype('uint8')
```

```python
# Step 5: Convert the image using the new gray level values
image_equalized = new_gray_values[image]
```

```python
# Display the original and equalized images along with their histograms
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
# Display the original image
axs[0, 0].imshow(image, cmap='gray')
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

# Display the equalized image
axs[0, 1].imshow(image_equalized, cmap='gray')
axs[0, 1].set_title('Equalized Image')
axs[0, 1].axis('off')

# Display the histogram of the original image
axs[1, 0].hist(image.flatten(), bins=256, range=[0, 256], color='blue')
axs[1, 0].set_title('Original Histogram')

# Display the histogram of the equalized image
axs[1, 1].hist(image_equalized.flatten(), bins=256, range=[0, 256], color='green')
axs[1, 1].set_title('Equalized Histogram')

plt.tight_layout()
plt.show()

```

```python
# Perform Histogram Stretching
min_intensity = np.min(image)
max_intensity = np.max(image)

# Stretching the original image
image_stretched = ((image - min_intensity) / (max_intensity - min_intensity)) * 255
image_stretched = image_stretched.astype('uint8')
```

```python
# Display the original and stretched images
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Display the original image
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

# Display the equalized image
axs[1].imshow(image_equalized, cmap='gray')
axs[1].set_title('Equalized Image')
axs[1].axis('off')

# Display the stretched image
axs[2].imshow(image_stretched, cmap='gray')
axs[2].set_title('Stretched Image')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
